```python
# Drop rows with any missing values
data.dropna(inplace=True)

# Format the dates
data.index = pd.to_datetime(data.index)
```

```python
# Calculate daily returns for each stock
daily_returns = data['Adj Close'].pct_change()

# Plot daily returns
daily_returns.plot(figsize=(10, 6))
plt.title("Daily Returns")
plt.xlabel("Date")
plt.ylabel("Returns")
plt.legend(tickers)
plt.show()
```

```python
# Calculate daily price change (volatility) for each stock
daily_volatility = data['Adj Close'].pct_change().std()

# Plot volatility
daily_volatility.plot(kind='bar', figsize=(8, 6))
plt.title("Daily Volatility")
plt.xlabel("Stock")
plt.ylabel("Volatility")
plt.show()
```

```python
# Plot stock prices over time
data['Adj Close'].plot(figsize=(10, 6))
plt.title("Stock Prices over Time")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend(tickers)
plt.show()
```

```python
# Plot histograms of daily price changes
daily_returns.hist(bins=50, figsize=(10, 6))
plt.title("Daily Price Changes (Volatility)")
plt.xlabel("Daily Returns")
plt.ylabel("Frequency")
plt.legend(tickers)
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = daily_returns.corr()

# Plot the correlation heatmap
plt.figure(figsize=(8, 6))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title("Correlation Heatmap")
plt.xticks(range(len(tickers)), tickers, rotation=45)
plt.yticks(range(len(tickers)), tickers)
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Function to fetch stock data and plot line chart
def plot_line_chart(ticker, start_date, end_date):
    # Fetch historical stock data from Yahoo Finance API
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    
    # Plot the line chart
    plt.figure(figsize=(10, 6))
    plt.plot(stock_data.index, stock_data['Close'], label=f"{ticker} Close Price", color='b')
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.title(f"{ticker} Stock Price over Time")
    plt.legend()
    plt.grid(True)
    plt.show()

# Define the stock tickers and date range
tickers = ["AAPL", "MSFT"]
start_date = "2022-01-01"
end_date = "2023-07-25"

# Plot line charts for each stock
for ticker in tickers:
    plot_line_chart(ticker, start_date, end_date)
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Function to fetch stock data and plot histogram
def plot_daily_price_changes(ticker, start_date, end_date):
    # Fetch historical stock data from Yahoo Finance API
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    
    # Calculate daily price changes (volatility)
    daily_returns = stock_data['Close'].pct_change()
    
    # Plot histogram of daily price changes
    plt.figure(figsize=(8, 6))
    plt.hist(daily_returns.dropna(), bins=50, color='b', alpha=0.7)
    plt.xlabel("Daily Price Changes")
    plt.ylabel("Frequency")
    plt.title(f"{ticker} Daily Price Changes Histogram")
    plt.grid(True)
    plt.show()

# Define the stock tickers and date range
tickers = ["AAPL", "MSFT"]
start_date = "2022-01-01"
end_date = "2023-07-25"

# Plot histograms of daily price changes for each stock
for ticker in tickers:
    plot_daily_price_changes(ticker, start_date, end_date)
```
